## 1.6 matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 6, 0.1)
y = np.sin(x)

plt.plot(x, y)
plt.show()

x = np.arange(0, 6, 0.1)
y1 = np.sin(x) #  <- 넘파이 배열
y2 = np.cos(x)
y3 = (y1 + y2) / 3

plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle="--", label="cos")
plt.plot(x, y3, linestyle=":", color='g', label="sin + cos / 3")
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin & cos')
plt.legend() # <- plot에서 이름 직접 전달해도 적어줘야되는듯
plt.show()
```
