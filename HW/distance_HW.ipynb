```python
import numpy as np
from math import *

def KL(p,q):
    return(sum(p*np.log(p/q)))

s1 = 'A girl is styling her hair'
s2 = 'A group of men play soccer on the beach'
s3 = 'One woman is measuring another woman\'s ankle'
s4 = 'A man is cutting up a cucumber'
s5 = 'A man is playing a harp'
s6 = 'A girl is brushing her hair'
s7 = 'A group of boys are playing soccer on the beach'
s8 = 'A woman measures another woman\'s ankle'
s9 = 'A man is slicing a cucumber'
s10 = 'A man is playing a keyboard'

l1 = [s1,s2,s3,s4,s5,s6,s7,s8,s9,s10]
l2 = []

for i in range(10):
    l1[i] = l1[i].split()

for i in range(10):
    l1[i] = sorted(l1[i])

for i in range(10):
    for j in range(len(l1[i])):
        if l1[i][j] in l2:
            continue
        else:
            l2.append(l1[i][j])

l2 = sorted(l2)


l3 = []
for i in range(10):
    l3.append(np.zeros(shape=(33,)))

alpha = 0.01

for i in range(10):
    for j in range(33):
        if l2[j] in l1[i]:
            l3[i][j] = 1
        l3[i][j] += alpha

for i in range(10):
    for j in range(10):
        print(round(KL(l3[i], l3[j]), 4))
    print()


```
